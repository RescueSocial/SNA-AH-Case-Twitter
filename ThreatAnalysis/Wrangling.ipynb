# Add the language column and the user.location to the data

```python
# create a list of columns to filter on 
cols = set(df_21.columns) | {'user.location', 'lang'}
```

```python
# Load the original full data
df_18 = pd.read_csv('tweets_2018.csv', lineterminator='\n')
df_19 = pd.read_csv('tweets_2019.csv', lineterminator='\n')
df_20 = pd.read_csv('tweets_2020.csv', lineterminator='\n')
df_21 = pd.read_csv('tweets_2021.csv', lineterminator='\n')
```

```python
# Filter on specific columns
df_18 = df_18[list(cols)]
df_19 = df_19[list(cols)]
df_20 = df_20[list(cols)]
df_21 = df_21[list(cols)]
```

```python
# merge all twitter data in one dataframe
df_merged = pd.concat([df_18, df_19, df_20, df_21])
print(df_merged.shape)
df_merged.head(1)
```

<br>

# Merged DataFrame Wrangling

```python
# rename columns
df_merged.rename(columns={'user.name': 'username',
                          'user.id_str': 'user_id',
                          'user.description': 'user_description',
                          'user.profile_banner_url': 'profile_banner_url',
                          'user.location': 'user_location',
                          'user.profile_image_url': 'profile_image_url',
                          'user.statuses_count': 'n_statuses',
                          'id_str': 'tweet_id',
                          'user.url': 'user_url',
                          'user.friends_count': 'n_friends',
                          'user.created_at': 'user_created_at',
                          'user.followers_count': 'n_followers',
                          'user.screen_name': 'user_screen_name',
                          'full_text': 'text'}, inplace=True);
```

```python
# change dtypes
df_merged['user_id'] = df_merged['user_id'].astype('str')
df_merged['tweet_id'] = df_merged['tweet_id'].astype('str')
df_merged["created_at"] = pd.to_datetime(df_merged["created_at"])
df_merged["user_created_at"] = pd.to_datetime(df_merged["user_created_at"])
```

```python
# Adding year, month, dayofmonth and a diff columns
df_merged["date"] = df_merged["created_at"].dt.date
df_merged["year"] = df_merged["created_at"].dt.year
df_merged["month"] = df_merged["created_at"].dt.strftime("%b")
df_merged["dayofmonth"] = df_merged["created_at"].dt.day
df_merged["diff"] = df_merged["created_at"] - df_merged["user_created_at"]
df_merged["days_diff"] = df_merged["diff"].dt.days
```
