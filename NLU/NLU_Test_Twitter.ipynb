```python
df0 = pd.read_csv('tweets_2019.csv')
print(df0.shape)
df0.head(1)
```

```python
df0.full_text.isnull().sum()
```

```python
df0.dropna(subset=['full_text'], inplace=True)
print(df0.shape)
df0.head(1)
```

## Save Data For Postman

```python
df0.rename(columns={'full_text': 'clean_text'}, inplace=True)
```

```python
# First we need to remove special caracters from text
df0.clean_text = df0.clean_text.apply(lambda x: x.replace('"', '').replace('\\','').replace('\t', ' ')\
                                                  .replace('\n', ' ').replace('\r', '').replace('       ',' ')\
                                                  .replace('      ',' ').replace('     ',' ').replace('   ',' ')\
                                                  .replace('  ',' '))
```

```python
# data1 = df0.clean_text.head(5000)
# data1.to_csv('nlp_data.csv', index=False)
```

## Load API OUTPUT (Postman)

```python
with open('Nlp Text classification test.txt', 'r') as f:
    response = f.read().replace('\n', '')
    
response = response[:-1]
response = '[' + response + ']'
nlp_list = ast.literal_eval(response)
```

```python
df1 = pd.DataFrame(nlp_list)
print(df1.shape)
df1.head()
```

```python
# Adding text classification to the comments dataframe
df2 = pd.concat([df1, df0.head(5000)], axis=1)
print(df2.shape)
df2.head()
```

https://stackoverflow.com/questions/27673231/why-should-i-make-a-copy-of-a-data-frame-in-pandas

```python
df_twitter = df2.copy()
```

```python
# rounding values to the nearest integer
df_twitter[['defense_AH', 'support_AH', 'offense_AH', 'defense_against_AH']] = \
df_twitter[['defense_AH', 'support_AH', 'offense_AH', 'defense_against_AH']].round()
df_twitter.head()
```

```python
# change the type of the rounded values to boolean
df_twitter['defense_AH'] = df_twitter['defense_AH'].astype('bool')
df_twitter['support_AH'] = df_twitter['support_AH'].astype('bool')
df_twitter['offense_AH'] = df_twitter['offense_AH'].astype('bool')
df_twitter['defense_against_AH'] = df_twitter['defense_against_AH'].astype('bool')
```

## Save The Classified DataFrame

```python
df_twitter.to_csv('twitter_classified.csv', index=False)
```
